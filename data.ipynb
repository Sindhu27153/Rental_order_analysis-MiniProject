```python
#checking for null values
df.isna().sum()
```

```python
#drop the null values
df.dropna(inplace=True)
```

```python
#check for duplicated values
df.duplicated().sum()
```

```python
#column names
df.columns
```

```python


#change the column name to lower case and Replace the columns with '_' for clarity and compatibility
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ",'_')
df.columns = df.columns.str.strip()
df.head()
```

```python
#add new colum "discount"
df["discount_price"] =round((df["list_price"]*df["discount_percent"])*0.01,2)
df["discount_price"]
```

```python
#add a new column to find a sale price 
df["sale_price"] = df["list_price"] -  df["discount_price"]
df
```

```python
#Add Profit column
df["profit"] = round(df["sale_price"] - df["cost_price"],2)
df
```

```python
orders_df = df[["order_id","order_date","product_id","ship_mode","segment",
                "country","city","state","postal_code","region"]]
orders_df
```

```python
orders_df.to_csv("orders.csv")
```
